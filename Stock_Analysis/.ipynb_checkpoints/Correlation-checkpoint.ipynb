# A simple script investigting the correlation between stocks, perhaps a useful feature for future use

```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats.stats import pearsonr
```

```python
file_path = "processed_data/aligned_dates/all_data_with_aligned_dates.npy"
```

```python
all_data = np.load(file_path)
num_to_symbol = np.load(file_path.replace(".npy","_DIC.npy"))
```

```python
n_entries, n_symbols, n_features = np.shape(all_data)
print("Dataset consists of {} entries, {} symbols and {} features".format(n_entries, n_symbols,n_features))
```

First normalize to index

```python
index_data = all_data.copy()
for symbol in range(n_symbols):
    for feature in range(4):
        index_data[:,symbol, feature] /= all_data[0,symbol, feature]
```

```python
plt.plot(all_data[:,0,3])
plt.xlabel("Time")
plt.ylabel("Index")
plt.grid()
plt.show()
```

```python
introduction_index = np.zeros(n_symbols, dtype=int)

for symbol in range(n_symbols):
    feature = 3 # Close values
    
    last_entry = n_entries-1
    
    last_value = index_data[last_entry,symbol,feature] # Sort out non-entry values
    
    while(index_data[last_entry-1,symbol,feature] == last_value): 
        last_entry -=1
    
    introduction_index[symbol] = last_entry
    
```

```python
correlation_matrix = np.zeros([n_symbols, n_symbols])
p_values = np.zeros(n_symbols)
feature = 3

for symbol in range(n_symbols):
    if symbol%50 == 0 : print("{}/{}".format(symbol, n_symbols))
        
    for comparing_symbol in range(symbol+1,n_symbols):
                
        least_common_entries = np.min([introduction_index[symbol], introduction_index[comparing_symbol]])
        
        time_series1 = index_data[:least_common_entries, symbol, feature]
        time_series2 = index_data[:least_common_entries, comparing_symbol, feature]
        
        rho, p = pearsonr(time_series1,time_series2 )
        
        if least_common_entries < 1000:
            p = 1
            rho = 0 
            
        p_values[symbol] = p
        correlation_matrix[symbol,comparing_symbol] = rho
    
        
```

```python
plt.imshow(correlation_matrix)
plt.colorbar()
```

```python
max_correlation = np.argmax(correlation_matrix[0,:])
```

```python
min_correlation = np.argmin(correlation_matrix[0,:])
```

```python
correlation_matrix[0,min_correlation]
```

```python
plt.plot(index_data[:1000,0,3],'r-')
plt.plot(index_data[:1000,max_correlation,3],'b-')
plt.plot(index_data[:1000,min_correlation,3],'g-')
plt.legend(['Stock','Highest Correlation Stock','Lowest Correlation Stock'])
```
