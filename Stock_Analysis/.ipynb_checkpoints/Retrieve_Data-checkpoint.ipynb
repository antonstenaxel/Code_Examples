# A script to retrieve large cap data from Swedish stocks using the Alpha Vantage API

```python
import requests, time, json
```

```python
API_key = "<API KEY>"
```

```python
with open("large_cap_symbols.txt") as f:
    symbols = []
    for line in f.readlines():
        
        symbol = line
        symbol = symbol.replace("\n","")
        symbol = symbol.replace(" ","-")
        symbol+=".ST"
        
        symbols += [symbol]
       
```

```python
def get_data(symbol):
    function = 'TIME_SERIES_DAILY_ADJUSTED'
    datatype = 'json'
    outputsize = 'full'
    r = requests.get('https://www.alphavantage.co/query?'+\
                'function='+function+\
                '&symbol='+symbol+\
                '&datatype='+datatype+\
                '&outputsize='+outputsize+\
                '&apikey='+API_key)
    
    data = r.json()
    
    if(len(data.keys()) == 1):
        raise Exception("ERROR: " + data['Error Message'])
    else:
        return data
```

```python
successes = 0
errors = 0
for symbol in symbols:
    try:
        data = get_data(symbol)
        successes += 1
        
        file_name = "stock_data/"+symbol.split(".")[0]+".json"
        with open(file_name, 'w') as outfile:
            json.dump(data, outfile)
            
        print("Success:",symbol)
    except Exception as e:
        errors += 1
        print("ERROR:",symbol)
    
   
    print("{}/{}".format(successes+errors,len(symbols)))
    
    # 5 calls/min limit 
    time.sleep(21)
```
